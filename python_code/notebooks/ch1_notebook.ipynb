####  Lists

```python
# define a list:
example_list = [1, 5, 41.3, 2.0]

# be careful with changes on variables pointing on example_list:
duplicate_list = example_list
duplicate_list[3] = 10000
print(f'duplicate_list: {duplicate_list}\n')
print(f'example_list: {example_list}\n')

# work on a copy of example_list:
example_list = [1, 5, 41.3, 2.0]
duplicate_list = example_list[:]
duplicate_list[3] = 10000
print(f'duplicate_list: {duplicate_list}\n')
print(f'example_list: {example_list}\n')
```

some notes:
- list index starts at zero
- can access a list entry or range of entries -- use start position and end position [start:end]
- this is what a list looks like
- to copy a list do list_name[:]
- it's be careful when altering a list -- work on a copy if you don't want the original to be altered<br>

```python
# defining a list: 
example2_list = [1, 2.3, 4, 5]

# showing an alteration of the original list
copy_ex2_list = example2_list
copy_ex2_list[2] = 700
print(f'duplicate list: {copy_ex2_list}')
print(f'example2 list (altered): {example2_list}\n')

# working on a copy instead
# first, reset the list:
example2_list = [1, 2.3, 4, 5]
copy_ex2_list = example2_list[:]
copy_ex2_list[2] = 700
print(f'duplicate list: {copy_ex2_list}')
print(f'example2 list: {example2_list}')
```

#### Dictionaries

Lists vs Dicts
- lists have an order; dictionaries are unordered
    - dict components are accessed by their unique keys

```python
# define and print a dict:
var1 = ['Florian', 'Daniel']
var2 = [96, 49]
var3 = [True, False]
example_dict = dict(name=var1, points=var2, passed=var3)
print(f'example_dict: \n{example_dict}\n')

# another way to define the dict:
example_dict2 = {'name': var1, 'points': var2, 'passed': var3}
print(f'example_dict2: \n{example_dict2}\n')

# get data type:
print(f'type(example_dict): {type(example_dict)}\n')

# access 'points':
points_all = example_dict['points']
print(f'points_all: {points_all}\n')

# access 'points' of Daniel:
points_daniel = example_dict['points'][1]
print(f'points_daniel: {points_daniel}\n')

# add 4 to 'points' of Daniel and let him pass:
example_dict['points'][1] = example_dict['points'][1] + 4
example_dict['passed'][1] = True
print(f'example_dict: \n{example_dict}\n')

# add a new variable 'grade':
example_dict['grade'] = [1.3, 4.0]

# delete variable 'points':
del example_dict['points']
print(f'example_dict: \n{example_dict}\n')
```

manually creating a dictionary <br>
you can access specific "columns" and values within the dict <br>
you can add a new variable with new values or remove a current variable

```python
# define and print a dict:
variable1 = ['Jordan', 'Tommi']
variable2 = [54, 63]
variable3 = [True, False]
ex_dict = dict(name=variable1, age=variable2, is_tall=variable3)
print(f'example dict:\n{ex_dict}')
# separator for each section of the chunk
print('===='*20)

# another way to define the dictionary:
ex_dict2 = {'name':variable1, 'age':variable2, 'is_tall':variable3}
print(f'example dict2: \n{ex_dict2}')
# separator 
print('===='*20)

# get data type:
print(f'example dict type: {type(ex_dict)}')
print('===='*20)

# access 'age':
age_all = ex_dict['age']
print(f'all ages: {age_all}')
print('===='*20)

# access whether or not Tommi is tall
tommi_height = ex_dict['is_tall'][1]
print(f'Tommi is tall? {tommi_height}')
print('===='*20)

# subtract 23 years from Jordan's age and make Tommi tall (he deserves it ;)
ex_dict['age'][0] = ex_dict['age'][0] -23
ex_dict['is_tall'][1] = True
print(f'new example dictionary: \n{ex_dict}')
print('===='*20)

# adding a new variable: 
ex_dict['favorite_sport'] = ['volleyball', 'basketball']
print('new dict:',ex_dict)
print('===='*30)

# removing a variable: 
del ex_dict['is_tall']
print('example dict:', ex_dict)
```
