### Python as a Calculator

```python
result1 = 1 + 1
print(f'result1: {result1}\n')

result2 = 5 * (4 - 1) ** 2
print(f'result2: {result2}\n')

result3 = [result1, result2]
print(f'result3: \n{result3}\n')
```

This is a basic example of a simple task that Python can do: acting as a calculator<br>
Can also use the print() function to output both text and a variable value<br>
This also shows that a list can be stored in a variable

```python
example1 = 2*2
print(f'example1 output: {example1}\n')

example2 = [example1, 'Some other variable']
print(f'example2: \n{example2}\n')
print('example2 is type:', type(example2))
```

### Module Math

```python
import math as someAlias

result1 = someAlias.sqrt(16)
print(f'result1: {result1}\n')

result2 = someAlias.pi
print(f'Pi: {result2}\n')

result3 = someAlias.e
print(f'Eulers number: {result3}\n')
```

module: a python file with code written on it already, so I can just utilize what's already been built<br>
Here's the math module which allow me to utilize "basic" math functions

```python
# already imported math so I'm not going to do it again
sqrt_example = someAlias.sqrt(100)
print(f'the square root of 100 is: {sqrt_example}\n')
# there's a pi function (honestly, I'm kinda curious as to the code behind this)
pi = someAlias.pi
print(f'Pi: {pi:.8}\n')     #cut off the decimal places as well
# there's e as well
e = someAlias.e
print(f'Euler\'s number: {e:.5}\n')
```
